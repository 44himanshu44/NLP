```python
import urllib.request
```

```python
link = "https://timesofindia.indiatimes.com/videos/city/delhi/no-respite-from-toxic-air-in-delhi-as-quality-remains-severe/videoshow/72081158.cms"
```

```python
req = urllib.request.Request(link)
response = urllib.request.urlopen(req)
the_page = response.read()
```

```python
the_page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(the_page, 'lxml')
```

```python
soup.find_all("p")[0].text
```

```python
soup.find_all("div")
```

```python
len(soup.find_all("p"))
```

```python
all_data = soup.find_all("p")
```

```python
for i in range(len(all_data)):
    print(all_data[i].text)
```

```python
text = all_data[4:-1]
```

```python
text = [text[i].text for i in range(len(text))]
```

```python
final_news = ' '.join(text)
final_news
```
